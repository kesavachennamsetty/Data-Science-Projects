```python
import pandas as pd
```

```python
import numpy as np
```

```python
df_train = pd.read_csv('train.csv')
```

```python
df_train = df_train[['Survived','Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']]
```

```python
df_train
```

```python
df_train['Embarked'].unique()
```

```python
df_train['Sex_int'] = df_train['Sex'].replace({'male':1,'female':0})
```

```python
df_train['Embarked_int'] = df_train['Embarked'].replace({'S':1,'C':2,'Q':3})
```

```python
df_train
```

```python
df_train = df_train.fillna(0)
```

```python
df_train_features = df_train[['Pclass','Age','SibSp','Parch','Fare','Sex_int','Embarked_int']]
```

```python
df_train_target = df_train['Survived']
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier(n_estimators = 200, random_state = 42)
```

```python
clf.fit(df_train_features,df_train_target)
```
